```python
import os
import requests
from typing import Dict, Any, Optional, List, Tuple
```

```python
KAKAO_NAVI_DIRECTIONS_URL = "https://apis-navi.kakaomobility.com/v1/directions"
```

```python
def kakao_car_directions(
    rest_api_key: str,
    origin: Tuple[float, float],
    destination: Tuple[float, float],
    priority: str = "RECOMMEND") -> Dict[str, Any]:
```

```python




class KakaoMobilityAPIError(RuntimeError):
    pass

def kakao_car_directions(
    rest_api_key: str,
    origin: Tuple[float, float],        # (lon, lat)
    destination: Tuple[float, float],   # (lon, lat)
    priority: str = "RECOMMEND",        # RECOMMEND / TIME / DISTANCE 등(문서 기준)
    waypoints: Optional[List[Tuple[float, float]]] = None,
    alternatives: bool = False,
    road_details: bool = False,
    summary: bool = True,
) -> Dict[str, Any]:
    """
    카카오모빌리티 자동차 길찾기 API 호출.
    반환: 응답 JSON(dict)
    """
    headers = {
        "Authorization": f"KakaoAK {rest_api_key}",
    }

    params = {
        "origin": f"{origin[0]},{origin[1]}",
        "destination": f"{destination[0]},{destination[1]}",
        "priority": priority,
        "alternatives": "true" if alternatives else "false",
        "road_details": "true" if road_details else "false",
        "summary": "true" if summary else "false",
    }

    # 경유지: "lon,lat|lon,lat|..." 형태로 전달(일반적으로 이렇게 사용)
    if waypoints:
        params["waypoints"] = "|".join([f"{lon},{lat}" for lon, lat in waypoints])

    resp = requests.get(KAKAO_NAVI_DIRECTIONS_URL, headers=headers, params=params, timeout=15)

    # 에러 처리(인증/쿼터/파라미터 오류 등)
    if not resp.ok:
        msg = f"HTTP {resp.status_code} - {resp.text}"
        raise KakaoMobilityAPIError(msg)

    return resp.json()

def extract_summary(route_json: Dict[str, Any]) -> Dict[str, Any]:
    """
    응답에서 거리/시간 등 요약을 뽑아주는 헬퍼.
    (응답 구조는 옵션/버전에 따라 달라질 수 있어 안전하게 get 사용)
    """
    routes = route_json.get("routes", [])
    if not routes:
        return {"has_route": False}

    s = routes[0].get("summary", {}) or {}
    # distance: m, duration: s 형태로 오는 경우가 많음
    distance_m = s.get("distance")
    duration_s = s.get("duration")

    return {
        "has_route": True,
        "distance_m": distance_m,
        "duration_s": duration_s,
        "distance_km": (distance_m / 1000) if isinstance(distance_m, (int, float)) else None,
        "duration_min": (duration_s / 60) if isinstance(duration_s, (int, float)) else None,
        "raw_summary": s,
    }

if __name__ == "__main__":
    # 1) 키는 코드에 하드코딩하지 말고 환경변수 추천
    # Windows(PowerShell):  setx KAKAO_REST_API_KEY "발급받은키"
    rest_key = os.getenv("KAKAO_REST_API_KEY")
    if not rest_key:
        raise SystemExit("환경변수 KAKAO_REST_API_KEY 를 설정해줘.")

    # 예시 좌표 (경도, 위도) — 필요에 맞게 바꿔서 사용
    origin = (127.10764191124568, 37.402464820205246)
    destination = (127.11056336672839, 37.39419693653072)
    waypoints = [(127.17354989857544, 37.36629687436494)]  # 경유지 1개 예시

    data = kakao_car_directions(
        rest_api_key=rest_key,
        origin=origin,
        destination=destination,
        priority="RECOMMEND",
        waypoints=waypoints,
        alternatives=False,
        road_details=False,
        summary=True,
    )

    info = extract_summary(data)
    print(info)
```
